```python
import pandas as pd
```

```python
list_of_orders_path = '/datalab/List_of_Orders_38EF37C2F3.csv'
order_details_path = '/datalab/Order_Details_F6D252B2DD.csv'
sales_target_path = '/datalab/Sales_target_1C8295CCDE.csv'
```

```python
list_of_orders = pd.read_csv(list_of_orders_path)
order_details = pd.read_csv(order_details_path)
sales_target = pd.read_csv(sales_target_path)
```

```python
list_of_orders.head(), order_details.head(), sales_target.head()
```

```python
merged_data = pd.merge(order_details, list_of_orders, on="Order ID")
```

```python
category_analysis = merged_data.groupby("Category").agg(Total_Sales=("Amount", "sum"),Average_Profit_Per_Order=("Profit", "mean"),Total_Profit=("Profit", "sum"),Total_Orders=("Order ID", "count"))
```

```python
category_analysis["Profit_Margin_Percentage"] = (category_analysis["Total_Profit"] / category_analysis["Total_Sales"] * 100)
```

```python
category_analysis
```

```python
category_analysis = category_analysis.reset_index()
```

```python
category_analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Total_Sales", data=category_analysis, palette="viridis")
plt.title("Total Sales by Category", fontsize=16)
plt.ylabel("Total Sales in rupees", fontsize=12)
plt.xlabel("Category", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Profit_Margin_Percentage", data=category_analysis, palette="coolwarm")
plt.title("Profit Margin Percentage by Category", fontsize=16)
plt.ylabel("Profit Margin (%)", fontsize=12)
plt.xlabel("Category", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Total_Orders",y="Total_Profit",hue="Category",size="Total_Sales",sizes=(100, 1000),data=category_analysis,palette="tab10")
```

```python
plt.title("Profit vs. Orders by Category", fontsize=16)
plt.xlabel("Total Orders", fontsize=12)
plt.ylabel("Total Profit ()", fontsize=12)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Total_Orders",y="Total_Profit",hue="Category",size="Total_Sales",sizes=(100, 1000),data=category_analysis,palette="tab10")
plt.title("Profit vs. Orders by Category", fontsize=16)
plt.xlabel("Total Orders", fontsize=12)
plt.ylabel("Total Profit ()", fontsize=12)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

Sales and Profitability Analysis

1).Merging the Data

We combined two datasets, one containing order information and the other detailing the specifics of each order, based on a common identifier: the Order ID. This allowed us to bring together customer and location data with the sales and profit information. The assumption was that every order ID in the order details dataset had a corresponding entry in the order list.

2).Key Findings

From the analysis, we observed noticeable differences in both sales and profitability across various categories. Heres a summary of the findings:

a).Clothing stands out with the highest profit margin of 8.03%. This is likely because the production and logistics costs are relatively lower for smaller items like apparel, and the high volume of orders boosts its profitability.

b).Electronics comes second in terms of total sales, but its profit margin is a bit lower at 6.35%. Higher-priced items in this category tend to offer better profit, though the logistics and handling costs still pose a challenge.

c).Furniture had the lowest profit margin at 1.81%. This can likely be attributed to the high shipping, storage, and handling costs associated with bulky items. Offering discounts to move large inventories might also be cutting into its profitability.
